## Activation Functions
### Visualizing Activation Functions
We'll visualize the effects of different activation functions.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

# Generate input data
x = np.linspace(-10, 10, 100)

# Plot activation functions
# Plot activation functions
plt.figure(figsize=(12, 10))

# Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)

# Tanh
plt.subplot(2, 2, 2)
plt.plot(x, tanh(x), label='Tanh')
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)

# ReLU
plt.subplot(2, 2, 3)
plt.plot(x, relu(x), label='ReLU')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)

# Leaky ReLU
plt.subplot(2, 2, 4)
plt.plot(x, leaky_relu(x), label='Leaky ReLU')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

## Loss Functions
### Calculating Loss Functions
We'll calculate common loss functions for a simple example.

````python
# Define true values and predictions
y_true = np.array([0, 1, 0, 1])
y_pred = np.array([0.1, 0.9, 0.2, 0.8])

# Mean Squared Error
mse = np.mean((y_true - y_pred) ** 2)
print(f'Mean Squared Error: {mse}')

# Cross-Entropy Loss
epsilon = 1e-15
y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
cross_entropy = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
print(f'Cross-Entropy Loss: {cross_entropy}')

# Hinge Loss
hinge_loss = np.mean(np.maximum(0, 1 - y_true * y_pred))
print(f'Hinge Loss: {hinge_loss}')
````
